# Iris Dataset

## Load libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set(color_codes=True)
%matplotlib inline
```

## Load dataset

```python
url = 'https://datahub.io/machine-learning/iris/r/iris.csv'
df = pd.read_csv(url)

# preview 5 first elements of df
df.head() 
```

## Basic Dataset investigation

```python
print("Shape of dataset:")
print(df.shape)
```

```python
print("Info on dataset:")
df.info()
```

```python
print("Descriptive statistics:")
df.describe()
```

```python
print("Unique classes:")
df['class'].unique()
```

```python
df.groupby('class').size()
```

```python
df.boxplot(by='class',figsize=(15,15))
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df,hue="class")
plt.show()
```
